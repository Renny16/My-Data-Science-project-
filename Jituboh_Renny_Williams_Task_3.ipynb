Merging Data

```python
import pandas as pd
# Merge orders and order_details
final_data = pd.merge(data4, data3, on="order_id", how="inner")

# Merge with pizzas
final_data = pd.merge(final_data, data, on="pizza_id", how="inner")

# Merge with pizza_types
final_data = pd.merge(final_data, data2, on="pizza_type_id", how="inner")
```

Observe Data

```python
#Check for overview of data
final_data.info()
```

```python
final_data.duplicated().sum()
```

```python
#check for null values
final_data.isna().sum()
```

Input the Correct Data types

```python
#convert date and time
final_data["date"]= pd.to_datetime(final_data["date"])
final_data["time"]= pd.to_datetime(final_data["time"])
```

```python
final_data.info()
```

Data Analysing

```python
#Total Revenue/Sales
final_data['revenue']= final_data['quantity']*final_data['price']
final_data['revenue'].sum()
```

```python
#Total Quantity Sold
final_data['quantity'].sum()
```

```python
#Total Order
final_data['order_id'].nunique()
```

```python
#How many Pizza types
final_data['pizza_type_id'].nunique()
```

```python
#Average Price of Pizzas
final_data['price'].mean()
```

```python
#peak hours of sales
final_data['hour'] = pd.to_datetime(final_data['time']).dt.hour
final_data.groupby('hour')['revenue'].sum().sort_values(ascending=False)
```

```python
#total sales by day of week

final_data['day'] = final_data['date'].dt.day_name()
final_data.groupby('day')['revenue'].sum().sort_values(ascending=False)
```

```python
#sales made in each Month
final_data['month'] = final_data['date'].dt.month_name()
final_data.groupby('month')['revenue'].sum()
```

   The sales trend shows noticeable monthly fluctuations rather than a steady increase or decrease. Revenue peaks in July, indicating strong mid-year performance. However, sales decline significantly in September and October, suggesting possible seasonal or consumer spending shifts. Overall, the business experiences moderate seasonality with stronger mid-year demand.

```python
final_data['month'] = final_data['date'].dt.month_name()
# Ensure month order
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_sales = final_data.groupby('month')['revenue'].sum().reindex(month_order)

plt.figure()
monthly_sales.plot(kind='bar')
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.title("Monthly Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
monthly_sales.plot(kind='line')
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.show()
```
